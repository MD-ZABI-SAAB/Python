# Logical Operators

1. True and false [AND OPERATOR]

- it alwys looks for false
- it's like multiplication of 0 & 1
- if both condition are satisfies dnn its true or its false
   
2. True or false  [OR OPERATOR]

- it always looks for true
- it's like addition of 1 & 0
- if one condition is satisfied it gives true or if both condition not satisfies its false

```python
# (and) Table

print  (True and False) 
print  (False and False)
print  (False and True)
print  (True and True)
```

```python
# (or) Table

print (True or False)
print (True or True)
print (False or False)
print (False or True)
```

```python
# Not Table

print(not True)
print(not False)
```

## Bitwise operator

- Bitwise and (&) -- it uses and table after converting it into binary
- Bitwise or (|) -- it uses or table after coverting it into binary
- Bitwise xor (^) --- it uses xor table after converting it into binary
- Bitwise left shift(<<) -- Use exponent
- Bitwise right shift(>>) -- use floor division
- Bitwise not (~) it uses not table

```python
a = 45
b = 21

print(a and b)
print(a & b)
```
