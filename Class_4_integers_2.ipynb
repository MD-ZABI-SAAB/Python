# Identity operators

- is and is not function
- checks the id (memory reusability) of variable(stack) and data type(hype)
- its is applicable for intigers within the range of -5 to 256 .
- not applicable for float
- not applicable for complex
- applicable for boolean
- applicable for none
- applicable for string but A-z , a-z , 0-9 , _ 

```python
a = 20
b = 20

print(id(a))
print(id(b))
print(a is b)   # Check the ID of a and b
```

```python
a = 20
b = 40

print(id(a))
print(id(b))
print(a is b)      # Check the ID of a and above code ID of a and B
```

```python
a = 257
b = 257

print(id(a))
print(id(b))
print(a is b)   # it checks the memory in the backend of python but these values are out of range so a new memory is created
print(a == b)   # it compares the value
```

```python
a = 45.99
b = 45.99

print(id(a))
print(id(b))
print(a is b)   # it creates a new memory
print(a == b)   # it compares the value
```

```python
a = True
b = True

print(id(a))
print(id(b))
print(a is b)   # it uses same memory
print(a == b)   # it compares the value
```

```python
a = "mohammed_zabiulla_777"
b = "mohammed_zabiulla_777"

print(id(a))
print(id(b))
print(a is b)  # It works as long as condition is satisfied
print(a == b)
```

```python
a = "mohammed_zabiulla@777"
b = "mohammed_zabiulla@777"

print(id(a))
print(id(b))
print(a is b)
print(a == b)
print(a>=b)
```

```python
a = None 
b = None

print(id(a))
print(id(b))
print(a is b)
print(a == b)
```

```python
a = "mohammed_zabiulla@777"
b = "mohammed_zabiulla@777"

print(id(a))
print(id(b))
print(a is not b)
```

```python
# Arithematiic Operator

a = 10.5
b = 20.5

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a**b)
print(a%b)
```

```python
# Comparison Operator

a = 10.5
b = 20.5

print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

```python
# Equlity Operator

a = 10.5
b = 20.5

print(a==b)
print(a!=b)
```

```python
# Logical Operator

a = 10.5
b = 20.5

print(a and b)
print(a or b)
print(not a != b)
print(not a == b)
```

```python
# Assignment Operations 

a = 10.5
b = 20.5
c = 30.5
d = 40.5
e = 50.5
f = 60.5
g = 70.5

a += b
print(a)

b -= a
print(b)

c *= b
print(c)

d /= c
print(d)

e %= d
print(e)

f //= e
print(f)

g **= f
print(g)
```

```python
# Identity Operatoe

a = 10.5
b = 20.5

print(a is b)
print(a is not b)
```
