# Data Types

1. Fundamental Datatypes (basic)
- int ---> any number without a decimal point
- float ---> any number with decimal point
- complex  ---> any number written in the form of a+bj where a is the real no and b is the imaginery no
- boolean  ---> decision making program either its true / false
- none ---> mostly used for misiing data it occupy the position but does not contain value 
- string ---> anything written inside ('..') ,( "..") ,( '''..''') ,( """...""") is known as string. inside a string u can write a word ,                                                                                       number , or a sentence
              * '..' is string but if it contains apost of es it will give error
              * ".." is string but it can print any statement even if it has apost of es
              *  '''..''' / """..""" it is used for multiple line 

2. Derived Datatypes (from the basic)
- list ---> anything written inside [..] seperated by , is is list
- tuple ---> anything written inside (..) seperated by , is tuple
- set  ---> anything written inside {..} seperated by , is set
- dict (dictionary) ---> anything written in {..} contains a key and value pair is dict

2. float = can be converted into int , complex , str ,bool

4. str()  = can be only converted into int or float or complex unless string contains matching values only
5. bool() = can be converted into int , float , complex , str

# Output Function

- Print (Plays an important role to display the output)

# Signature: print(*args, sep=' ', end='\n', file=None, flush=False)

- Docstring: Prints the values to a stream, or to sys.stdout by default.

1. sep: string inserted between values, default a space.

2. end: string appended after the last value, default a newline.
   
4. file: a file-like object (stream); defaults to the current sys.stdout.
   
6. flush: whether to forcibly flush the stream.
Type:      builtin_function_or_method

```python
a = 10
b = 20

print(a)
print(b)
```

```python
print(a,b)                  # as u can see the output has space 
```
