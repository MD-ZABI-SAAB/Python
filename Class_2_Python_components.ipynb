# Data Types

A. Fundamental Data types
 - intiger = any no without decimal point 
 - float = any no with decimal point
 - Complex = any no written in form of a+bj where a is a real no and b is imaginery no
 - Bool = either its true / false
 - None = used for missing data
 - String = anything written inside '' , " " , ''' ''' , """ """

B. Derived Data types
 - List = anything written in []
 - Tuple = anything written in()
 - Set = anytthing written in {}
 - Dictionary = it is writttten in {} but it has an key and its value

# Derived Data Types

```python
# each element written within [] and seperated by , is a list
# A list is a mustable dattype

a = [1,2,3,4,5]
a[0] = 0.1   
print(a,type(a))
```

```python
# each element written within () and seperated by , is a tuple
# A tuple is an immutable datatype

b = (1,2,3,4,5) 
print(b,type(b))
```

```python
# each element written within {} and seperated by , is a set
# A set is a collection of unique values

c = {1,2,2,2,2,2,2,2,3,4,5} 
print(c,type(c))
```

```python
# anytthing which is written in {} containing key and its value
# key and its value is referred to as single item

d = {'name' : 'xyz' , 'age' : 25 , 'salary' : 50000}
print(d, type(d))
```

# For clear understanding dict

```python
a = ['xyz' , 35 , 1000000]

print('name =', a[0])
print('age =', a[1])
print('salary =', a[2])
```

```python
# now lets say the name came at the end 
a = [35 , 1000000 , 'xyz']

print('name =', a[0])
print('age =', a[1])
print('salary =', a[2]) # as u can see we have jumbled our values 
```

```python
# now lets try the same thing with dictionary

a = {'name' : 'xyz' ,'age' : 35 , 'salary' : 1000000}

print('name =', a['name'])
print('age =', a['age'])
print('salary =', a['salary']) # here we dont give the posittion we provide the key to extract the data
```

```python
a = {'age' : 35 , 'salary' : 1000000 , 'name' : 'xyz'}

print('name =', a['name'])
print('age =', a['age'])
print('salary =', a['salary'])
# EVEN IF WE CHANGE THE POSITION OF A DATAPOINT WE CAN STILL EXTRACT IN CORRECT ORDER
```

# Typecasting

- convert one data type to another

1. int() = can be converted into float , complex , str ,bool
2. float = can be converted into int , complex , str ,bool
3. complex() = cannot be converted into int or float
4. str()  = can be only converted into int or float or complex unless string contains matching values only
5. bool() = can be converted into int , float , complex , str

```python
# syntax 1

a = float(64)
print(a ,type(a))
```

```python
# syntax 2

a = 64
print(a,type(float(a)))
```

```python
b = int(24.5)
print(b,type(b))
```

```python
c = complex(25+25.5)
print(c,type(c))
```

```python
d = str(33)
print(d,type(d))
```

```python
e = bool(99999)
print(e,type(e))
```

```python
f = complex('46+55j')
print(f,type(f))
```
